# Variáveis Qualitativas:

As variáveis qualitativas podem também ser chamadas de categóricas, são relacionadas a qualidade e características dos indivíduos,não podem ser medidas númericamente e geralmente está relacionada a classificação dos indivíduos. São divididas em dois grupos, qualitativas nominais e Ordinais.


 



[texto do link](https:// [texto do link](https:**Qualitativas Nominais:**

---


- não representam ordenação, apenas características dos indivíduos, como cor dos olhos, sexo,não fumante, fumante, etc.

Exemplos:

*No caso abaixo trago como exemplo o dataset Iris, ele contém dados de largura, altura e das petalas e das sepalas e o target que no caso seia o tipo de iris que pode ser setosa, versicolor e virginica,
image.pngesses dados são muito utilizados para problemas de classificação em Machine learning..


Nesse caso os valores contidos na coluna 'Tipos de Iris' se enquadram na questão de variáveis qualitativas, onde descrevem o tipo de iris dadas as caractetísticas....*

```python
from sklearn.datasets import load_iris
import pandas as pd
iris = load_iris()
columns = list(iris.feature_names)
df = pd.DataFrame(iris.data,columns=columns)

df['Tipos de Iris'] = iris.target
df['Tipos de Iris'] = df['Tipos de Iris'].replace(to_replace= [0, 1, 2], value = ['setosa', 'versicolor', 'virginica'])

df.head()
```

Qualitativas ordinais:
- São dados categóricos que possuem algum tipo de ordenação, exemplo estágio de doenças, classificação em uma corrida(1º lugar,2ºlugar, etc..)e também podemos considerar como dados categóricos ordinais meses do ano, como utilizo no exemplo abaixo:

# Variáveis Quantitativas  

As variáveis quantitativas são características que podem ser descritas por números, sendo estas classificadas entre contínuas e discretas.

- Quantitativas contínuas:

Ainda utilizando o exemplo do dataset Iris, variáveis quantitativas contínuas, são aquelas avaliadas em números com casas decimais( em programação float, double), geralmente oriundos de alguma medição, exemplo peso, valores de imóveis e outros. Nesse caso poderiamos classificar como variáveis desse tipo as colunas/features :  

'sepal length (cm)',  
 'sepal width (cm)',  
 'petal length (cm)',  
 'petal width (cm)'  
 Que nesse caso representam o tamanho da petala ou sepala em centímetros...

- Quantitativa discretas:   


*Ao analisar as variáveis quantitativas discretas, podemos dizer que se referem a númeeros inteiros resultados de contagem, número de pessoas por metro quadrado, número de filhos, quantidade de população de uma cidade. Mas tranzendo a explicação para onosso exemplo, poderiamos então criar uma nova tabela, onde fariamos a contagem de quantas plantas foram analisadas de cada tipo, nessa nova tabela teriamos o campo espécie e quantidade observada, neste caso teriamos como variavel quantitativa discreta os valores da feature/coluna "quantidade observada", segue o exemplo abaixo*:

```python
tipos=df['Tipos de Iris'].value_counts().to_dict()
df_especies =pd.DataFrame()
df_especies["especies"]=tipos.keys()
df_especies["quantidade observada"]=tipos.values()
df_especies.head()
```
