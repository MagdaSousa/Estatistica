Neste caso estamos lidando com intervalos de classe, então ao calcular medidas de tendencia central elas sofrem algumas modificações:

Formula da média em dados agrupados  
image.png

```python
import pandas as pd
```

```python
#Criando um dataframe com os dados da tarefa
media_em_intervalos = pd.DataFrame(columns=['Classes','Frequencia absoluta(fi)','Media das classes(xi)','fi*xi'])

#populando a tabela com os dados#
media_em_intervalos["Classes"]=[" 1000.00 |-- 1500.00"," 1500.00 |-- 2000.00"," 2000.00 |-- 2500.00"," 2500.00 |-- 3000.00"," 3000.00 |-- 3500.00"]
media_em_intervalos["Frequencia absoluta(fi)"]=[15,22,30,18,15]
media_em_intervalos.head(10)
```

Calculando a média entre dos intervalos e inserindo na tabela

```python
#biblioteca em python que faz calculos estatísticos
import statistics
```

```python
#este trecho : statistics.mean([1000.00, 1500.00]), está chamando a função mean, da biblioteca statistics e calculando a média entre o inicio e o final do intervalo de cada classe

media_em_intervalos.at[0, 'Media das classes(xi)'] = statistics.mean([1000.00, 1500.00])
media_em_intervalos.at[1, 'Media das classes(xi)'] = statistics.mean([1500.00 ,2000.00])
media_em_intervalos.at[2, 'Media das classes(xi)'] = statistics.mean([2000.00, 2500.00])
media_em_intervalos.at[3, 'Media das classes(xi)'] = statistics.mean([2500.00,3000.00])
media_em_intervalos.at[4, 'Media das classes(xi)'] = statistics.mean([3000.00,3500.00])
media_em_intervalos.head(10)
```

Multiplicar a frequencia absoluta pela média de cada intervalo:

```python
media_em_intervalos['fi*xi'] = media_em_intervalos["Frequencia absoluta(fi)"] * media_em_intervalos['Media das classes(xi)']

media_em_intervalos.head(10)
```

Formula da média para interválos de classe:
Média Aritmética = ∑(fi * xi) / ∑fi


```python
#soma da frequencia absoluta ∑fi

soma_frequencia_absoluta=media_em_intervalos["Frequencia absoluta(fi)"].sum()
soma_frequencia_absoluta
```

```python
#soma total do resultado da multiplicação entre a frequencia absoluta e a média das classes "fi*xi"

soma_fi_xi=media_em_intervalos["fi*xi"].sum()
soma_fi_xi
```

Calculando média para intervalos de classe

```python
media_aritimetica_para_intervalos_de_classe =soma_fi_xi/soma_frequencia_absoluta
media_aritimetica_para_intervalos_de_classe
```
